```python
x1 = np.linspace(0,10,1000)
y = np.dot(x1*x1,5)
```

```python
#np.shape(y)
```

```python
plt.plot(x1,y,".")
plt.show()
```

```python
np.shape(y)
```

```python
#Let N be the number of points
N = 800
i = 0+1j
#k = np.linspace(0,10,N-1)
```

```python
#w = np.sum(2*np.pi*k)
```

```python
final = []
complex_expo = 0
#np.array(complex_expo)
np.array(final)
for k in range(0,1000):
    for n in range(0,1000):
        complex_expo = complex_expo + y[k]*np.exp(-(2*np.pi*i*k*n)/N)
    final = np.append(final,complex_expo)
```

```python
print(np.shape(final))
```

```python
fig, ax = plt.subplots()
plt.plot(final.real, final.imag)
```

```python
back = []
inv_complex_expo = 0
np.array(back)
for k in range(0,1000):
    for n in range(0,1000):
        inv_complex_expo =inv_complex_expo + final[k]*np.exp((2*np.pi*i*k*n)/N)
    back = np.append(back,inv_complex_expo)
```

```python
np.shape(back)
```

```python
fig2, bx = plt.subplots()
plt.plot(back.real, back.imag)
```
