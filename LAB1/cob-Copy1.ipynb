```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
import mpld3
mpld3.enable_notebook()
```

```python
x1 = np.linspace(0,10,1000)
y = np.dot(x1*x1,5)
f = y + x1*1j
#Let N be the number of points
total = 1000
percent = int(input("The percentage of points to be considered: "))
N = int((percent/100)*total)

complex_expo = [[0 for a in range(1000)] for b in range(1000)]
np.array(complex_expo)
inv_complex_expo = [[0 for a in range(N)] for b in range(N)]
np.array(inv_complex_expo)
for k in range(0,1000):
    for n in range(0,1000):
        complex_expo[k][n] =np.exp(-2j*np.pi*n*k/len(y))
for k in range(0,N):
    for n in range(0,N):
        inv_complex_expo[k][n] =np.exp(2j*np.pi*n*k/N)
sign = np.dot(f,complex_expo)
recon = (np.dot(inv_complex_expo,sign))/N
```

```python
plt.plot(f.real,f.imag)
```

```python
plt.plot(sign.real,sign.imag)
```

```python
plt.plot(recon.real,recon.imag)
```
